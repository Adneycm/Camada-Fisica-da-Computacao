```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.fftpack import fft, fftshift
```

As funções a seguir servirão de apoio para construção de senoides e obtenção da transformada discreta de Fourier

```python
def generateSin(freq, time, fs):
    n = time*fs #numero de pontos
    x = np.linspace(0.0, time, n)  # eixo do tempo
    s = np.sin(freq*x*2*np.pi)
    return (x, s)
```

```python
def calcFFT(signal, fs):
    # https://docs.scipy.org/doc/scipy/reference/tutorial/fftpack.html
    #y  = np.append(signal, np.zeros(len(signal)*fs))
    N  = len(signal)
    T  = 1/fs
    xf = np.linspace(-1.0/(2.0*T), 1.0/(2.0*T), N)
    yf = fft(signal)
    return(xf, fftshift(yf))
```

# Prática em Fourier

## Senoide única

 - Gere uma senoide de 5Hz. 
 - Exiba o sinal no tempo e o Fourier do mesmo (módulo e fase). Lembre-se de que a transformada retorna uma lista de complexos!
 - Analíse a composição espectal desse sinal.


```python
# Geração de uma senoide com frequência de 5Hz durante 4 segundos e com uma frequência de amostragem de 200 pontos 
x, s = generateSin(5, T, fs)

plt.figure(figsize=(25,10))
plt.plot(x,s)
plt.xlabel("Tempo (s)", fontsize = 18)
plt.ylabel("Amplitude", fontsize = 18)
plt.show()

xf, fftshift_yf = calcFFT(s, fs)

plt.figure(figsize=(25,10))
plt.plot(xf,np.abs(fftshift_yf))
plt.xlabel("Frequência (Hz)", fontsize = 18)
plt.ylabel("Amplitude", fontsize = 18)
plt.axvline(x=5, ymin=0, ymax=400, color='r')
plt.xlim(0,10)
plt.show()

print(f"xf = {xf[:10]}\n\n")
print(f"fftshift_yf = {fftshift_yf[:10]}")
```

## Senoides 

 - Gere duas senoide de mesma amplitude e frequência 5Hz e 8Hz
 - Exiba o sinal no tempo e o Fourier do mesmo (módulo)
 - Analíse a composição espectal desse sinal.
 


```python
x_5Hz, s_5Hz = generateSin(5, T, fs)
x_8Hz, s_8Hz = generateSin(8, T, fs)

plt.figure(figsize=(25,10))
plt.plot(x_5Hz, s_5Hz, label="Senoide de f=5Hz")
plt.plot(x_8Hz, s_8Hz, label="Senoide de f=8Hz")
plt.plot(x_8Hz, s_8Hz+s_5Hz, label="Soma das senoides")
plt.xlabel("Tempo (s)", fontsize = 18)
plt.ylabel("Amplitude", fontsize = 18)
plt.legend(loc='upper right', fontsize=18)
plt.show()

xf, yf = calcFFT(s_5Hz+s_8Hz, fs)
#xf_8Hz, yf_8Hz = calcFFT(s_8Hz, fs)

plt.figure(figsize=(25,10))
plt.plot(xf,np.abs(yf))
plt.xlabel("Frequência (Hz)", fontsize = 18)
plt.ylabel("Amplitude", fontsize = 18)
plt.axvline(x=5, ymin=0, ymax=400, color='r')
plt.axvline(x=8, ymin=0, ymax=400, color='r')
plt.xlim(0,10)
plt.show()
```
